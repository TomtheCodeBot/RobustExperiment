```python
y = [float(x) for x in df.columns[1:]]
```

```python
dataframe = [[]  for _ in range(len(df["noise_type\intensity"]))]
for i in df.columns[1:]:
    for k in range(len(df["noise_type\intensity"])):
        #dataframe[k].append(float(df[i][k].split("%")[0]))
        dataframe[k].append(df[i][k]*100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
  

# plot lines
plt.plot( y,dataframe[0], label = df["noise_type\intensity"][0],marker='x')
plt.plot( y,dataframe[1], label = df["noise_type\intensity"][1],marker='*')
plt.plot( y,dataframe[2], label = df["noise_type\intensity"][2],marker='o')
plt.plot( y,dataframe[3],'--', label = df["noise_type\intensity"][3],marker='o')
#plt.plot( y,dataframe[4],'--', label = df["noise_type\intensity"][4],marker='x')
#plt.plot( y,dataframe[5],'--', label = df["noise_type\intensity"][5],marker='*')
plt.xlabel('Noise intensity',fontsize=14) 
plt.ylabel('Accuracy (%)',fontsize=14) 
plt.tick_params(axis='both', which='major', labelsize=14)

# displaying the title
plt.legend(loc = "lower left",fontsize=12)
plt.show()
```
