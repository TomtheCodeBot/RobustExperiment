```python
import pandas as pd
```

```python
df = pd.read_excel('/home/ubuntu/RobustExperiment/IMDB_0.1_scale_clean_accuracy_results.xlsx')  
df
```

```python
y = [float(x) for x in df.columns[1:]]
```

```python
dataframe = [[]  for _ in range(len(df["noise_type\intensity"]))]
for i in df.columns[1:]:
    for k in range(len(df["noise_type\intensity"])):
        dataframe[k].append(float(df[i][k].split("%")[0]))
```

```python
import matplotlib.pyplot as plt
import numpy as np
  

# plot lines
plt.plot( y,dataframe[0], label = df["noise_type\intensity"][0],marker='o')
plt.plot( y,dataframe[1], label = df["noise_type\intensity"][1],marker='o')
plt.plot( y,dataframe[2], label = df["noise_type\intensity"][2],marker='o')
plt.plot( y,dataframe[3], label = df["noise_type\intensity"][3],marker='o')
plt.plot( y,dataframe[4], label = df["noise_type\intensity"][4],marker='o')
plt.plot( y,dataframe[5], label = df["noise_type\intensity"][5],marker='o')
plt.xlabel('Noise intensity') 
plt.ylabel('clean Accuracy (%)') 
  
# displaying the title
plt.title("IMDB BERT 0.1-Interval Noise Clean Accuracy")
plt.legend()
plt.show()
```
